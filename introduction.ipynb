# Penguins and Islands

We will import the library and name it "sns" so we have to type less. Then we will use a function called `load_dataset` to load a table of measurements of different penguins and assign that to a variable called "pengu" to use it later.

```python
import seaborn as sns

pengu = sns.load_dataset("penguins")
pengu
```

As you can see this has data for 344 penguins like their species, where they live, body measurements and their sex. 

*BTW, "bill" is the name for a penguin's beak.*

We might expect that heavier penguins have longer flippers. To see if that is true we might want to plot the weight against the flipper length. This can be done using the `scatterplot` function from seaborn.

```python
sns.scatterplot(x="body_mass_g", y="flipper_length_mm", data=pengu)
```

Cool, that shows that our hypothesis seems to be correct. But what about different islands? Let's also color the points by the island each penguin came from.

```python
sns.scatterplot(x="body_mass_g", y="flipper_length_mm", hue="island", data=pengu)
```

Okay, looks like our hypothesis still works but the Biscoe island definitely has larger penguins than the other two islands.

So weight and flipper length are definitely correlated but we can also check that with math by calculating the correlation coefficients. Turns out that our `pengu` dataset comes with its own set of functions. One of them is called `corr` and calculates the Pearson correlation coefficients between all pairs of columns in the data.


```python
pengu.corr()
```

From that we can see that weight and flipper length are positively correlated. We can even fit lines to the data in the plot. For that, we only have to substitute the `scatterplot` functions with the `lmplot` function (short for "linear model plot"). Lets also make the plot a bit larger by using the `height` argument.

```python
sns.lmplot(x="body_mass_g", y="flipper_length_mm", hue="island", data=pengu, height=10)
```

The plot now shows linear regression lines for each island. The shaded areas denote standard error of the line (the uncertainty associated with the linear model). As we can see the slope of the line is a bit steeper for the Biscoe island, meaning that the penguins there have longer flippers for a particular weight compared to the other two islands. Pretty neat, right?
